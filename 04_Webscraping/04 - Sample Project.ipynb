# Libraries

```python
from bs4 import BeautifulSoup, SoupStrainer

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

from pprint import pprint
import time
```

```python
options = webdriver.ChromeOptions()
## if you run this code locally and want to see how Selenium operates the borwser, comment the following lines
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

## if you run this code locally and want to see how Selenium operates the borwser, uncomment the following line
#options.add_argument("--window-size=1920x1080")
```

# Google Image Captioning

## Load Page

```python
driver = webdriver.Chrome('chromedriver', options=options)
driver.get('https://www.google.com/imghp?hl=en')
```

## Accept

```python
try :
  el = driver.find_element(by=By.XPATH, value="//button[2]")
  el.click()
except:
  el = driver.find_element(by=By.XPATH, value="//button[1]")
  el.click()
```

## Camera upload button

```python
 # Find cam button
el = driver.find_element(by=By.XPATH, value='//div[@aria-label="Search by image"]')
el.click()
```

## Upload Image

<img src='https://images.unsplash.com/photo-1535224206242-487f7090b5bb' width='100%'/>

```python
url = 'https://images.unsplash.com/photo-1535224206242-487f7090b5bb'
el = driver.find_element(by=By.XPATH, value='//input[@name="image_url"]')
el.send_keys(url)
el.send_keys(Keys.RETURN)
```

## Get Search Text

```python
el = driver.find_element(by=By.XPATH, value='//input[@name="q"]')

soup = BeautifulSoup(driver.page_source,
                     parse_only=SoupStrainer('input', attrs={'name': 'q'}))

soup.input['value']
```
