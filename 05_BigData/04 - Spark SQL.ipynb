## Read via PySpark

```python
from pyspark import SparkConf, SparkContext
from pyspark.sql import SQLContext

conf = SparkConf()\
        .setAppName("Spark-Core")\
        .setMaster("local[*]")\
        .set("spark.mongodb.input.uri", "mongodb://127.0.0.1:27017/eurostat.directors")\
        .set("spark.mongodb.output.uri", "mongodb://127.0.0.1:27017/eurostat.directors")\
        .set("spark.jars.packages", "org.mongodb.spark:mongo-spark-connector_2.12:3.0.1")

sc = SparkContext(conf=conf).getOrCreate()
sqlc = SQLContext(sc)
```

```python
rdd = sqlc.read.format("com.mongodb.spark.sql.DefaultSource").option("uri", "mongodb://127.0.0.1:27017/eurostat.directors").load()
```

```python
rdd.show()
```
